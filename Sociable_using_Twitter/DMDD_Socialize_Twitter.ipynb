```python
# defining variables for search

new_search = "#socialize"
num_tweets = 500
hashtag_tweets = []
user_info = []
```

```python
#scraping twitter to get most recent tweets that have "#socialize" 

tweets = tweepy.Cursor(api.search_tweets,q=new_search,count=num_tweets,lang="en",tweet_mode='extended')
#extracting the tweet_info and user_info of all "#socialize" tweets
for tweet in tweets.items(500):
    if not tweet.entities["urls"]:
        continue
    else:
        hashtag_tweets.append((tweet.id, tweet.user.id, tweet.full_text, tweet.entities["urls"][0]["url"],tweet.retweet_count, tweet.favorite_count, tweet.created_at))
        user_info.append((tweet.user.id, tweet.user.screen_name, tweet.user.location, tweet.user.created_at))
    
```

```python
#creating a dataframe to store user information

df_user = pd.DataFrame(user_info, columns =['user_id', 'username', 'location', 'created_at'])
df_user
```

```python
DRIVER_NAME = 'SQL Server'
SERVER_NAME = 'SHRISHAILYA1919'
DATABASE_NAME = 'SHRI'

connection_string = f"""
DRIVER={{{DRIVER_NAME}}};
SERVER={SERVER_NAME};
DATABASE={DATABASE_NAME};
TRUST_Connection=yes;
"""


#connect to sql server

connection = odbc.connect(connection_string)

cursor = connection.cursor()

print(cursor)
```

```python

# SQL command to create Tweets table in the database

cursor.execute("DROP TABLE IF EXISTS Tweets")

sql_command = """CREATE TABLE Tweets (
Twitter_handle VARCHAR(10),
Tweet_text VARCHAR(20),
image_url VARCHAR(200),
description VARCHAR(100),
Twitter_id INT,
Created_date DATE,
PRIMARY KEY (Twitter_id)
);
"""
 
# execute the statement
cursor.execute(sql_command)

```

```python
# SQL command to create Tweet_tag table in the database

cursor.execute("DROP TABLE IF EXISTS Tweet_tag")

sql_command = """
CREATE TABLE Tweet_tag (
Tweettag_id INT,
Twitter_id INT,
tag VARCHAR(200),
PRIMARY KEY (Tweettag_id)
);

"""
 
# execute the statement
cursor.execute(sql_command)
```

```python
# SQL command to create Tweet_mentions table and add constraint in the database

cursor.execute("DROP TABLE IF EXISTS Tweet_mentions")

sql_command = """
CREATE TABLE Tweet_mentions (
Twitter_id INT,
Tweet_mentionsId INT,
src_user VARCHAR(200),
trg_user VARCHAR(100),
PRIMARY KEY (Tweet_mentionsId)
);

ALTER TABLE Tweet_mentions
ADD CONSTRAINT Twitter_id FOREIGN KEY (Tweets)
REFERENCES Tweets(Twitter_id);

"""
 
# execute the statement
cursor.execute(sql_command)
```

```python
# SQL command to create User table and add constraint in the database

cursor.execute("DROP TABLE IF EXISTS User")

sql_command = """
CREATE TABLE User (
Twitter_id INT,
name VARCHAR(20),
profile_image_url VARCHAR(200),
description VARCHAR(100),
Followers INT,
Following INT
);

ALTER TABLE User
ADD CONSTRAINT Twitter_id FOREIGN KEY (Tweets)
REFERENCES Tweets(Twitter_id);

"""
 
# execute the statement
cursor.execute(sql_command)
```

```python
# SQL command to create Socialize_account  table in the database

cursor.execute("DROP TABLE IF EXISTS Socialize_account")

sql_command = """
CREATE TABLE Socialize_account (
Twitter_id INT,
password VARCHAR(20),
Socialize_id INT,
PRIMARY KEY (Socialize_id)
);

"""
 
# execute the statement
cursor.execute(sql_command)
```

```python
# SQL command to create Match and add constraint  table in the database

cursor.execute("DROP TABLE IF EXISTS Match")

sql_command = """
CREATE TABLE Match (
Interest_Category INT,
Messages VARCHAR(255),
User1_id INT,
User2_id INT,
Following INT,
Match_id INT
PRIMARY KEY (Match_id)
);

ALTER TABLE Match
ADD CONSTRAINT User1_id FOREIGN KEY (Tweets)
REFERENCES Tweets(Twitter_id);

ALTER TABLE Match
ADD CONSTRAINT User2_id FOREIGN KEY (Tweets)
REFERENCES Tweets(Twitter_id);

ALTER TABLE Match
ADD CONSTRAINT Interest_Category FOREIGN KEY (Interest_Category)
REFERENCES Interest_Category(Interest_Category_id);

"""
 
# execute the statement
cursor.execute(sql_command)
```

```python
# SQL command to create Interest_Category table and add constraint in the database

cursor.execute("DROP TABLE IF EXISTS Interest_Category ")

sql_command = """
CREATE TABLE Interest_Category (
Interest_Category_id INT,
Twitter_id INT,
Category_id INT,
PRIMARY KEY (Interest_Category_id)
);

ALTER TABLE Interest_Category
ADD CONSTRAINT Twitter_id FOREIGN KEY (Tweets)
REFERENCES Tweets(Twitter_id);

ALTER TABLE Interest_Category
ADD CONSTRAINT Category_id FOREIGN KEY (Category)
REFERENCES Category(Category_id);

"""
 
# execute the statement
cursor.execute(sql_command)
```

```python
# SQL command to create Category table in the database

cursor.execute("DROP TABLE IF EXISTS Category ")

sql_command = """
CREATE TABLE Category (
Category_id INT,
name VARCHAR(20),
PRIMARY KEY (Category_id)
);

"""
 
# execute the statement
cursor.execute(sql_command)
```
